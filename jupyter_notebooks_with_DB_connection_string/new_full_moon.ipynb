```python
import pandas as pd
import pathlib as path
import datetime as dt
```

```python
moonchart_df = pd.read_csv('MoonChart2009.csv')
moonchart_df
```

```python
moonchart_df.info()
```

```python
moonchart_df.drop('Time', axis=1, inplace=True)
moonchart_df
```

```python
# Convert Moon Type to boolean
# Create copy of pandas DataFrame
moon_bool = moonchart_df.copy()
# Replace string by boolean
moon_bool['Moon Type'] = moon_bool['Moon Type'].map({'Full': True, 'New': False})
print(moon_bool) 
```

```python
moonchart_df = pd.DataFrame(moon_bool)
moonchart_df
```

```python
moonchart_df.rename(columns= {'Date':'Moon_Date', 'Moon Type':'Full_Moon'}, inplace=True)
moonchart_df
```

```python
# convert column to datetime pandas
moonchart_df['Moon_Date'] = pd.to_datetime(moonchart_df['Moon_Date'])
moonchart_df.info()
```

```python
# date in MM-DD-YYYY format
moonchart_df['Moon_Date'] = moonchart_df['Moon_Date'].dt.strftime('%m-%d-%Y')
moonchart_df
```

```python
moonchart_df.to_csv('Boolean_MoonChart_2009.csv', index=False)
```
